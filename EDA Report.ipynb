```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Store.csv")
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Plot 1: Profit distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Profit'], bins=50, kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Sales vs Profit scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', alpha=0.6)
plt.title('Sales vs Profit by Product Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Average Profit by Region
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Region', y='Profit', estimator=np.mean)
plt.title('Average Profit by Region')
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Profit by Discount Bins
bins = pd.cut(df['Discount'], bins=[-0.01, 0.1, 0.2, 0.3, 0.5, 1.0])
df['Discount Bin'] = bins
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Discount Bin', y='Profit')
plt.title('Profit Distribution by Discount Levels')
plt.xlabel('Discount Bin')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# Plot 5: Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Plot 6: Category Count Plot
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Category')
plt.title('Order Count by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plot 7: Cross-tab heatmap (Segment vs Region)
crosstab = pd.crosstab(df['Segment'], df['Region'])
plt.figure(figsize=(8, 5))
sns.heatmap(crosstab, annot=True, fmt='d', cmap='Blues')
plt.title('Customer Segment vs Region')
plt.tight_layout()
plt.show()
```
